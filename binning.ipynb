## Importing the dataset

```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv('cleaned_exams.csv')
dataframe
```

## Binning

```python

bins = [0, 72, 76, 80, 84, 88, 92, 96, 101]
labels = ['Failure', 'Passing', 'Below Average', 'Average', 'Above Average', 'Good', 'Superior', 'Excellent']

math_labels = pd.cut(dataframe['math score'], bins=bins, labels=labels, right=False)
reading_labels = pd.cut(dataframe['reading score'], bins=bins, labels=labels, right=False)
writing_labels = pd.cut(dataframe['writing score'], bins=bins, labels=labels, right=False)

dataframe.insert(7, 'math grade', math_labels)
dataframe.insert(9, 'reading grade', reading_labels)
dataframe.insert(11, 'writing grade', writing_labels)

dataframe
```

```python
dataframe.to_csv('graded_exams.csv', index=False)
```
